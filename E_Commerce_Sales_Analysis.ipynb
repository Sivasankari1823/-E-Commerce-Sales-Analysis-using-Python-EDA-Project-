```python

# Check for duplicates
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
if not duplicate_rows.empty:
    display(duplicate_rows.head())


#  Extract month and year into new columns
df['month'] = df['order_date'].dt.month
df['year'] = df['order_date'].dt.year

# Optional: Extract full month name and weekday
df['month_name'] = df['order_date'].dt.strftime('%B')
df['weekday'] = df['order_date'].dt.day_name()

#Preview result
print(df[['order_date', 'year', 'month', 'month_name', 'weekday']].head())


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
numerical_cols = ['age', 'unit_price', 'quantity', 'total_price', 'shipping_fee']

for col in numerical_cols:
    print(f"\n--- {col.upper()} ---")
    print(df[col].describe())
    print(f"Skewness: {df[col].skew():.2f}")

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(df[col], kde=True, ax=ax[0], color='skyblue')
    ax[0].set_title(f'{col} Distribution')

    sns.boxplot(x=df[col], ax=ax[1], color='salmon')
    ax[1].set_title(f'{col} Boxplot')
    plt.tight_layout()
    plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
categorical_cols = ['gender', 'region', 'category', 'shipping_status', 'returned']

for col in categorical_cols:
    print(f"\n{col.upper()} Frequency:")
    print(df[col].value_counts())
    print(df[col].value_counts(normalize=True) * 100)

    sns.countplot(x=col, data=df, order=df[col].value_counts().index, palette='pastel')
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
sns.barplot(x='category', y='unit_price', data=df, estimator='mean', palette='Set2')
plt.title('Average Unit Price by Category')
plt.show()
```

```python
sns.barplot(x='region', y='total_price', data=df, estimator='sum', palette='coolwarm')
plt.title('Total Sales by Region')
plt.show()
```

```python
sns.barplot(x='gender', y='quantity', data=df, estimator='mean', palette='muted')
plt.title('Avg Quantity by Gender')
plt.show()
```

```python
return_rate = df[df['returned'] == 'Yes'].groupby('category').size() / df.groupby('category').size() * 100
sns.barplot(x=return_rate.index, y=return_rate.values, palette='Reds')
plt.title('Return Rate by Category (%)')
plt.ylabel('Return %')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
for col in numerical_cols:
    skew_val = df[col].skew()
    print(f'{col}: Skewness = {skew_val:.2f}  {"Right" if skew_val > 0.5 else "Left" if skew_val < -0.5 else "Normal"} skewed')
```

```python
def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower) | (data[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
    return outliers

# Check for each
for col in numerical_cols:
    detect_outliers_iqr(df, col)
```

```python
#  Filter data for 2024 and 2025
sales_2024 = df[df['year'] == 2024].groupby('month')['total_price'].sum()
sales_2025 = df[df['year'] == 2025].groupby('month')['total_price'].sum()

# Ensure all months are represented (1-12) even if some months have 0 sales
all_months = range(1, 13)
sales_2024 = sales_2024.reindex(all_months, fill_value=0)
sales_2025 = sales_2025.reindex(all_months, fill_value=0)

# Plot side-by-side trend comparison
plt.figure(figsize=(10, 5))
plt.plot(sales_2024.index, sales_2024.values, marker='o', linestyle='-', label='2024', color='green')
plt.plot(sales_2025.index, sales_2025.values, marker='s', linestyle='--', label='2025', color='blue')

# Customize the chart
plt.title('Total Sales by Month (2024 vs 2025)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(1, 13))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
